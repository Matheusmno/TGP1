```python
# Required Libraries
import numpy as np
import pandas as pd
import textwrap

#from pyBibX.base import pbx_probe
from google.colab import data_table
import datetime
from datetime import datetime
import glob
import pickle

from google.colab.data_table import DataTable
DataTable.max_columns = 100
```

```python
df = pd.read_csv("/content/drive/MyDrive/Projeto CNI - Produções/Dados Energia/Sustainable Aviation Fuel - SAF/SAF_Patents.txt", sep=r'\t', engine='python')
df_ipc_desc = pd.read_excel("/content/drive/MyDrive/Projeto CNI - Produções/Dados Energia/Sustainable Aviation Fuel - SAF/Dicionário IPC/IPC4_Desc.xlsx")
df_ipc_desc.Description = df_ipc_desc.Description.str.capitalize()
df_ipc_desc.rename(columns={"Description": "IPC4_Desc"}, inplace=True)
```

```python
patent_data = {
    'PN ': 'Patent Number',
    ' TI ': 'Title',
    ' AU ': 'Authors or Inventors',
    ' AE ': 'Patent Assignee',
    ' GA ': 'IDS Number',
    ' AB ': 'Abstract / BHTD Critical Abstract',
    ' TF ': 'Technology Focus Abstract',
    ' EA ': 'Early access date; Equivalent Abstract, Editor Address',
    ' DC ': 'Derwent Class Code(s)',
    ' MC ': 'Major Concepts or Derwent Manual Code(s)',
    ' IP ': 'International Patent Classification',
    ' PD ': 'Publication Date; Patent Details',
    ' AD ': 'Application Details and Date',
    ' FD ': 'Further Application Details',
    ' PI ': 'Publisher City; Patent Priority Information',
    ' DS ': 'Designated States',
    ' FS ': 'Field of Search',
    ' CP ': 'Cited Patent(s)',
    ' CR ': 'Cited References',
    ' DN ': 'DCR Number',
    ' MN ': 'Markush Number',
    ' RI ': 'ResearcherIDs; Ring Index Number',
    ' CI ': 'Derwent Compound Number',
    ' RG': 'Derwent Registry Number'
}
```

```python
df.rename(columns=patent_data, inplace=True)
df
```

```python
df['International Patent Classification'] = df['International Patent Classification'].str.split(";  ")
df['Patent Assignee'] = df['Patent Assignee'].str.split(";  ")
df
```

```python
# Function to set color based on condition
def set_color_br(value):
    if value.startswith('BR'):
        return 'green'
    elif "BR" in value:
        return 'yellow'
    else:
        return 'blue'
```

```python
# Apply the function to create a new column
df['circle_color'] = df['Patent Number'].apply(set_color_br)
df
```

```python
df_exploded = df.explode("Patent Assignee")
df_exploded = df_exploded.explode("International Patent Classification")
df_exploded.reset_index(inplace=True)
```

```python
# Function to set color based on condition
def set_color_univ(value):
    if value.startswith('UNIV') or "MASSACHUSETTS INST TECHNOLOGY (MASI-C)" in value:
        return 'green'
    else:
        return 'blue'
```

```python
df_exploded['circle_color'] = df_exploded['Patent Assignee'].apply(set_color_univ)
df_exploded
```

```python
df_exploded["International Patent Classification"] = df_exploded["International Patent Classification"].str[:4]
```

```python
df_exploded = pd.merge(df_exploded, df_ipc_desc, how="left", left_on="International Patent Classification", right_on="IPC4")
```

```python
def extract_Matrix(text):
    import re
    pattern = r'\(([^)]+)\)$'  # Regex pattern
    match = re.search(pattern, text)  # Search for the pattern in the text
    if match:
        return match.group(1)  # Return the captured group
    else:
        return None  # Return None if no match is found

df_exploded['Patent Assignee Matrix'] = df_exploded['Patent Assignee'].apply(extract_Matrix)
```

```python
# Calcula o tamanho da família de patentes
# Usar filter para excluir ID "WO"
#df_exploded["Family Size"] = df_exploded["Patent Number"].str.split(";  ").apply(lambda x : len(x))
df_exploded["Family Size"] = df_exploded["Patent Number"].str.split(";  ").apply(lambda x : len(list((filter(lambda x : not x.startswith("WO"), x)))))
```

```python
# Exclui as famílias de patentes apenas com WO
df_wo = df_exploded[df_exploded["Family Size"] == 0]
df_exploded = df_exploded[df_exploded["Family Size"] > 0]
```

```python
agora = datetime.now()
df_exploded["Patent Year"] = df_exploded["Application Details and Date"].str.split("    ").apply(lambda row: int(row[2].split(";   ")[0].split(" ")[-1]))
df_exploded["Years until expiration"] = (df_exploded["Patent Year"] + 20) - agora.year

df_wo["Patent Year"] = df_wo["Application Details and Date"].str.split("    ").apply(lambda row: int(row[2].split(";   ")[0].split(" ")[-1]))
df_wo["Years until expiration"] = (df_wo["Patent Year"] + 2.5) - agora.year   # 30 months

df_exploded
```

```python
def get_patent_color(value):
    if value <= 5:
        return "rgba(0, 255, 0, 0.9)"  # Light green with 50% opacity
    elif value <= 10:
        return "rgba(144, 238, 144, 0.8)"  # Light green with 50% opacity
    elif value <= 15:
        return "rgba(240, 128, 128, 0.8)"  # Light coral with 50% opacity
    else:
        return "rgba(255, 0, 0, 0.5)"  # Red with 50% opacity

def get_patent_color_wo(value):
    if value <= 0.5:
        return 'grey'
    else:
        return 'purple'

# Apply the function to create the fading red hexadecimal column
df_exploded["triangle_color"] = df_exploded["Years until expiration"].apply(get_patent_color)
df_wo["triangle_color"] = df_wo["Years until expiration"].apply(get_patent_color_wo)
df_wo
```

```python
df_derwent = df_exploded[["index", "Patent Number", "Title", "Patent Assignee", "Patent Assignee Matrix", "circle_color", "triangle_color", "International Patent Classification", "IPC4_Desc", "Family Size"]]
df_wo = df_wo[["index", "Patent Number", "Title", "Patent Assignee", "Patent Assignee Matrix", "circle_color", "triangle_color", "International Patent Classification", "IPC4_Desc", "Family Size"]]
```
